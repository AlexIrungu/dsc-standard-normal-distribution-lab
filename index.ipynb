## Define the interval bounds which contain 99% of the observations   

> **Hint**: Recall the empirical rule related to $3\sigma$.

```python
# Perform any calculations necessary here
# Assuming mean (mu) and standard deviation (sigma) are known
mu = yield_data['yield'].mean()
sigma = yield_data['yield'].std()

# Calculate the interval bounds
lower_bound = mu - 3*sigma
upper_bound = mu + 3*sigma

```

```python
# Write your answer here 
print(f"The interval bounds containing approximately 99% of the observations are: [{lower_bound}, {upper_bound}]")
```

## Compute and interpret the z-score for a tree yielding 35 pounds of apples

```python
# Calculate z
# Given value
X = 35

# Calculate mean and standard deviation
mu = yield_data['yield'].mean()
sigma = yield_data['yield'].std()

# Compute z-score
z_score = (X - mu) / sigma

print(f"The z-score for a tree yielding 35 pounds of apples is: {z_score}")
```

```python
# Interpret the result

### Interpretation
"""- **Z-score > 0**: The yield is above the average yield.
- **Z-score < 0**: The yield is below the average yield.
- **Z-score = 0**: The yield is exactly at the average."""

# The magnitude of the z-score indicates how far and in what direction (above or below) the yield is from the mean, measured in terms of standard deviations.
```

## Suppose a tree has a z-score of 1.85. Interpret this z-score. What is the yield of this tree?

```python
# Interpret the z score
"A z-score of 1.85 indicates that the yield of this tree is 1.85 standard deviations above the mean yield of all trees. This is a relatively high yield, suggesting that this tree is significantly more productive than the average tree."

```

```python
# Calculate yield
# Given z-score
z = 1.85

# Assuming mean (mu) and standard deviation (sigma) are known
mu = yield_data['yield'].mean()
sigma = yield_data['yield'].std()

# Calculate the yield
X = mu + z * sigma
```

```python
# What is the yield ?
print(f"The yield of the tree with a z-score of 1.85 is: {X} pounds")
```

##  Convert each tree’s yield to a z-score so the new variable is the “z-score for weight”

The units are still the apple trees. For the data set of all z-scores:

* What is the shape? 
* The mean? 
* The standard deviation?

```python
# Give your solution here 
# Calculate mean and standard deviation of the original yields
mu = yield_data['yield'].mean()
sigma = yield_data['yield'].std()

# Convert each yield to a z-score
yield_data['z-score for weight'] = (yield_data['yield'] - mu) / sigma

# Calculate and print the shape, mean, and standard deviation of the z-scores
print(f"Shape of the z-score dataset: Same as original")
print(f"Mean of z-scores: {yield_data['z-score for weight'].mean()}")
print(f"Standard deviation of z-scores: {yield_data['z-score for weight'].std()}")
```

```python
import matplotlib.pyplot as plt

# Plotting the histogram of the z-scores
plt.hist(yield_data['z-score for weight'], bins=30, alpha=0.7, color='blue')
plt.title('Distribution of Z-scores for Apple Yields')
plt.xlabel('Z-score for Weight')
plt.ylabel('Frequency')

# Show the plot
plt.show()
```
